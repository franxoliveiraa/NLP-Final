### Analyze contents of movie clusters

```python
## Read in info for movies in each cluster

clusters = []
current = []
num_clusters = 30
count = 0

with open(('clusters_' + str(num_clusters)), 'r') as fread:
    for line in fread.readlines():
        line = line.strip()
        if "CLUSTER" in line:
            count += 1
            print(count)
            if current:
                clusters.append(current)
                current = []
            continue
        if line:
            list = []
            line = line.replace('[','').replace(']','').replace('\n','').split(', ')
            genre = ' '.join(line[5:len(line)])
            list = line[0:5]
            list.append(genre)
            current.append(list)
    clusters.append(current)
```

```python
## Report composition of each cluster by decade

decades = ["1930s", "1940s", "1950s", "1960s", "1970s", "1980s", "1990s", "2000s", "2010s"]

for i in range(len(clusters)):
    print("\nCLUSTER {n}".format(n=i))
    for decade in decades:
        total = 0
        count = 0
        for movie in clusters[i]:
            if movie[4].replace("'", "") == decade:
                count += 1
            total += 1
        print(decade + " {p}%".format(p=(count/total)*100))
                
```

```python
## Report composition of each cluster by genre

genres = ['Comedy', 'Drama', 'Fantasy', 'Action', 'Adventure', 'Horror', 'Crime', 'Thriller', 'Mystery', 'Romance', 'History', 'War', 'Western', 'Film-Noir', 'Sci-Fi', 'Biography', 'Family', 'Animation', 'Adult', 'Music']

for j in range(len(clusters)):
    print("\nCLUSTER {n}".format(n=j))
    for genre in genres:
        total = 0
        count = 0
        for movie in clusters[j]:
            if genre in movie[5]:
                count += 1
            total += 1
        print(genre + " {p}%".format(p=(count/total)*100))
        
```

```python
# Read in movie vectors from file

import numpy as np

movievectors = []
with open('movie_vectors', 'r') as fread:
    file = fread.read().rstrip()
    vectors = file.replace("[", "").strip().split(']')
    for i in range(4050):
        current = np.zeros(100)
        values = vectors[i].split()
        for j in range(100):
            current[j] = current[j] + float(values[j].strip('\n'))
        movievectors.append(current)
print("READ " + str(len(movievectors)) + " MOVIE VECTORS\n")
```

```python
## Cluster movie vectors (30 clusters)

from sklearn.cluster import KMeans

num_clusters = 30
kmmovies = KMeans(n_clusters=num_clusters, random_state=0)
movieclusters = kmmovies.fit_predict(movievectors)
centers = kmmovies.cluster_centers_
```

```python
## Read in metadata for movies in sample

import pandas as pd

sample_metadata = []

sample_file = pd.read_excel("sample_movie_metadata.xlsx", sheet_name='Sample')
sample_df = pd.DataFrame(sample_file, columns=['index','textID','title','year','decade','genre'])
sample_movie_indices = sample_df['index'].tolist()
sample_movie_indices = [int(i) for i in sample_movie_indices]
ids = sample_df['textID'].tolist()
titles = sample_df['title'].tolist()
decades_data = sample_df['decade'].tolist()
years = sample_df['year'].tolist()
genres = sample_df['genre'].tolist()
for i in range(len(sample_movie_indices)):
    index = sample_movie_indices[i]
    id = ids[i]
    title = titles[i]
    decade = decades_data[i]
    year = years[i]
    genre = genres[i]
    meta = [index, id, title, year, decade, genre]
    sample_metadata.append(meta)

indices = [int(i) for i in range(len(decades_data))]
index_to_decade = dict(zip(indices,decades_data))
index_to_year = dict(zip(indices,years))
index_to_vector = dict(zip(indices,movievectors))
index_to_cluster = dict(zip(indices,movieclusters))
index_to_genre = dict(zip(indices,genres))
index_to_title = dict(zip(indices,titles))
print("READ METADATA\n")
```

```python
## Get distance from each movie to each cluster

movieclusterdistances = kmmovies.transform(movievectors)

cluster_movies = []
cluster_movie_vectors = []
cluster_movie_decades = []
cluster_movie_clusters = []
cluster_movie_years = []
cluster_movie_genres = []
```

```python
## Find 5 movies closest to each cluster

for i in range(num_clusters):
    closest = []
    for j in range(len(movieclusterdistances)):
        current = (float(movieclusterdistances[j][i]),j)
        if closest:
            if len(closest) == 5:
                if current[0] < closest[-1][0]:
                    closest[-1] = current
                    closest.sort(key = lambda x: x[0])
            else:
                closest.append(current)
                closest.sort(key = lambda x: x[0])
        else:
            closest.append(current)
    cluster_movies.append(closest)
print(cluster_movies)
```
